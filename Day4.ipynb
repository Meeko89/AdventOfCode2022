```julia
function read_sections(filename)
    content = []
    f = open(filename)
    
    for line in readlines(f)
        section1 = string(split(line, ",")[1])
        section1_split = [parse(Int64, string(split(section1, "-")[1])), parse(Int64, string(split(section1, "-")[2]))] 
        section2 = string(split(line, ",")[2])
        section2_split = [parse(Int64, string(split(section2, "-")[1])), parse(Int64, string(split(section2, "-")[2]))] 
        line_content = [section1_split, section2_split]
        push!(content, line_content)
    end
    return content
end
```

```julia
function sort_on_length(elf_pair)
    sorted_pair = [[0,0], [0,0]]
    if elf_pair[1][2] - elf_pair[1][1] < elf_pair[2][2] - elf_pair[2][1]
        # println(typeof(sorted_pair[1]), " ", typeof(elf_pair[1]))
        sorted_pair[1] = elf_pair[2]
        sorted_pair[2] = elf_pair[1]
        # println("Sorted ", sorted_pair)
    else
        sorted_pair = elf_pair
        # println("Not Sorted ", sorted_pair)
    end
    return sorted_pair
end
```

```julia
# Day 4 Part 1
puzzle_input = "data/day4_data.txt"
# puzzle_input = "data/day4_test.txt"
sections = read_sections(puzzle_input)

result = 0
for elf_pair in sections
    elf_pair = sort_on_length(elf_pair)
    # println("Pair ", elf_pair)
    # println("Type ", typeof(elf_pair))
    # println("First ", elf_pair[1][1], ", ", elf_pair[2][1])
    # println("Second ", elf_pair[1][length(elf_pair[1])], ", ", elf_pair[2][length(elf_pair[2])])
    if elf_pair[1][1] <= elf_pair[2][1] &&
        
        elf_pair[1][length(elf_pair[1])] >= elf_pair[2][length(elf_pair[2])]
        result = result + 1
        println(elf_pair, " match")
    else 
        println(elf_pair, " -----")
    end
end
print(result)
```

```julia
function sort_on_start(elf_pair)
    sorted_pair = [[0,0], [0,0]]
    if elf_pair[1][1] > elf_pair[2][1]
        # println(typeof(sorted_pair[1]), " ", typeof(elf_pair[1]))
        sorted_pair[1] = elf_pair[2]
        sorted_pair[2] = elf_pair[1]
        # println("Sorted ", sorted_pair)
    else
        sorted_pair = elf_pair
        # println("Not Sorted ", sorted_pair)
    end
    return sorted_pair
end
```

```julia
# Day 4 part 2
puzzle_input = "data/day4_data.txt"
# puzzle_input = "data/day4_test.txt"
sections = read_sections(puzzle_input)

result = 0
for elf_pair in sections
    elf_pair = sort_on_start(elf_pair)
    # println("Pair ", elf_pair)
    # println("Type ", typeof(elf_pair))
    # println("First ", elf_pair[1][1], ", ", elf_pair[2][1])
    # println("Second ", elf_pair[1][length(elf_pair[1])], ", ", elf_pair[2][length(elf_pair[2])])
    if elf_pair[1][2] >= elf_pair[2][1]
        result = result + 1
        println(elf_pair, " match")
    else 
        println(elf_pair, " -----")
    end
end
print(result)
```
