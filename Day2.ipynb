```julia
function read_rounds(filename)
    content = []
    f = open(filename)
    
    for line in readlines(f)
        round = [Int(Char(line[1])) - Int('A') + 1, Int(Char(line[3])) - Int('X') + 1]
        push!(content, round)
    end
    return content
end
```

```julia
function get_round_result(elf_shape, my_shape)
    result = elf_shape - my_shape
    if result == 0
        # Draw
        return my_shape + 3 
    elseif result == 1 || result == -2
        # Loss
        return my_shape+ 0
    elseif result == 2 || result == -1
        # Win
        return my_shape + 6
    end
end
```

```julia
# Day 2 part 1
# all_rounds = read_rounds("data/day2_test.txt")
all_rounds = read_rounds("data/day2_data.txt")
tot_result = 0
for round in all_rounds
    tot_result = tot_result + get_round_result(round)
end
println(tot_result)
```

```julia
function get_round_shapes(round_info)
    round_result = round_info[2]
    elf_shape = round_info[1]
    WIN = 3
    DRAW = 2
    LOSS = 1
    if round_result == DRAW
        my_shape = elf_shape
    elseif round_result == LOSS
        if elf_shape == 1
            my_shape = 3
        elseif elf_shape == 2
            my_shape = 1
        else
            my_shape = 2
        end
    else
        if elf_shape == 1
            my_shape = 2
        elseif elf_shape == 2
            my_shape = 3
        else
            my_shape = 1
        end
    end

    round = [elf_shape, my_shape]
    return round
end
```

```julia
# Day 2 part 2
# all_rounds = read_rounds("data/day2_test.txt")
all_rounds = read_rounds("data/day2_data.txt")
tot_result = 0
for round_info in all_rounds
    round = get_round_shapes(round_info)
    tot_result = tot_result + get_round_result(round)
end
println(tot_result)
```
