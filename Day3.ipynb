```julia
function read_rugsacks(filename)
    content = []
    f = open(filename)
    
    for line in readlines(f)
        line_content = String(line)
        push!(content, line_content)
    end
    return content
end
```

```julia
function get_priority(item)
    if item >= 'A' && item <= 'Z'
        priority = Int(item) - Int('A') + 27
    else
        priority = Int(item) - Int('a') + 1 
    end
    return priority
end
```

```julia
# Day 3 part 1
# rugsacks = read_rugsacks("data/day3_test.txt")
rugsacks = read_rugsacks("data/day3_data.txt")
shared_items_priority = []
for rugsack in rugsacks
    compartment1 = SubString(rugsack, 1, Int(length(rugsack)/2))
    compartment2 = SubString(rugsack, Int(length(rugsack)/2)+1, length(rugsack))
    shared_items = []
    for item in compartment1 
        if occursin(item, compartment2) && !(get_priority(item) in shared_items)
            push!(shared_items, get_priority(item))
        end
    end
    println(compartment1, " ", compartment2)
    println(shared_items)
    append!(shared_items_priority, shared_items)
end
println(shared_items_priority)
println(sum(shared_items_priority))
```

```julia
# Day 3 part 2
# rugsacks = read_rugsacks("data/day3_test.txt")
rugsacks = read_rugsacks("data/day3_data.txt")
groups_items_priority = []
for group_start in 1:3:length(rugsacks)
    group_item = "."
    for item in rugsacks[group_start]
        if occursin(item, rugsacks[group_start + 1]) && occursin(item, rugsacks[group_start + 2])
            group_item = item
            break
        end
    end
    push!(groups_items_priority, get_priority(group_item))
end
println(groups_items_priority)
println(sum(groups_items_priority))
```
