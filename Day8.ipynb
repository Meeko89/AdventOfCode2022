```julia
function read_tree_map(filename)
    content = []
    f = open(filename)
    
    for line in readlines(f)
        line_content = [parse(Int64, x) for x in line]
        push!(content, line_content)
    end
    return content
end
```

```julia
function is_visible(row, col, tree_map)
    # Check if tree is on edge
    if row - 1 < 1 || row + 1 > length(tree_map) ||
         col - 1 < 1 || col + 1 > length(tree_map[1])
        return true
    end

    print("[", row, ", ", col, "]")
    # Check row visibility
    visible_directions = 4
    for i in 1:row-1
        if tree_map[i][col] >= tree_map[row][col]
            visible_directions -= 1
            print(" u")
            break
        end
    end
    for i in row+1:length(tree_map)
        if tree_map[i][col] >= tree_map[row][col]
            visible_directions -= 1
            break
            print(" d")
        end
    end

    # Check column visibility
    for j in 1:col-1
        if tree_map[row][j] >= tree_map[row][col]
            visible_directions -= 1
            print(" l")
            break
        end
    end

    for j in col+1:length(tree_map[1])
        if tree_map[row][j] >= tree_map[row][col]
            visible_directions -= 1
            print(" r")
            break
        end
    end

    println(" (", visible_directions, ")")
    return visible_directions > 0
end
```

```julia
function get_visible_trees(tree_map)
    visible_trees = 0
    for row in 1:length(tree_map)
        for column in 1:length(tree_map[1])
            if is_visible(row, column, tree_map)
                visible_trees += 1
            end
        end
    end
    return visible_trees
end
```

```julia
# Day 8 part 1

input_file = "data/day8_data.txt"
# input_file = "data/day8_test.txt"

tree_map = read_tree_map(input_file)
# println(tree_map)

visible_trees = get_visible_trees(tree_map)
print("Visible Trees: ", visible_trees)
```

```julia
function get_scenic_score(row, col, tree_map)

    # Check row visibility
    viewing_distance = 1
    
    tree_view = 0
    for i in row-1:-1:1
        tree_view += 1
        if tree_map[i][col] >= tree_map[row][col]
            break
        end
    end
    viewing_distance *= tree_view
    tree_view = 0

    for i in row+1:length(tree_map)
        tree_view += 1
        if tree_map[i][col] >= tree_map[row][col]
            break
        end
    end
    viewing_distance *= tree_view
    tree_view = 0

    # Check column visibility
    for j in col-1:-1:1
        tree_view += 1
        if tree_map[row][j] >= tree_map[row][col]
            break
        end
    end
    viewing_distance *= tree_view
    tree_view = 0

    for j in col+1:length(tree_map[1])
        tree_view += 1
        if tree_map[row][j] >= tree_map[row][col]
            break
        end
    end
    viewing_distance *= tree_view
    tree_view = 0

    return viewing_distance
end
```

```julia
function get_best_scenic_tree(tree_map)
    best_scenic_tree = 0
    for row in 1:length(tree_map)
        for column in 1:length(tree_map[1])
            scenic_score = get_scenic_score(row, column, tree_map)
            best_scenic_tree = max(best_scenic_tree, scenic_score)
        end
    end
    return best_scenic_tree
end
```

```julia
# Day 8 part 1

input_file = "data/day8_data.txt"
# input_file = "data/day8_test.txt"

tree_map = read_tree_map(input_file)
# println(tree_map)

best_scenic_tree = get_best_scenic_tree(tree_map)
print("Best Scenic Tree: ", best_scenic_tree)
```
